```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("./data/fungut_metadata.tsv", sep='\t')

# Convert relevant columns to numeric where needed
df['age_years_sample'] = pd.to_numeric(df['age_years_sample'], errors='coerce')
df['bmi_sample'] = pd.to_numeric(df['bmi_sample'], errors='coerce')
df['latitude_sample'] = pd.to_numeric(df['latitude_sample'], errors='coerce')
df['longitude_sample'] = pd.to_numeric(df['longitude_sample'], errors='coerce')

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")

# Plot 1: Age distribution by sex
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age_years_sample', hue='sex_sample', multiple="stack", kde=True)
plt.title('Age Distribution by Sex')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()

# Plot 2: BMI distribution by diet type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='diet_type_sample', y='bmi_sample')
plt.xticks(rotation=45, ha='right')
plt.title('BMI Distribution by Diet Type')
plt.xlabel('Diet Type')
plt.ylabel('BMI')
plt.show()

# Plot 3: Geographical distribution of samples (latitude vs. longitude)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='longitude_sample', y='latitude_sample', hue='country_sample', palette='tab20', s=100)
plt.title('Geographical Distribution of Samples')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Plot 4: IBD prevalence by diet type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='diet_type_sample', hue='ibd_sample')
plt.xticks(rotation=45, ha='right')
plt.title('IBD Prevalence by Diet Type')
plt.xlabel('Diet Type')
plt.ylabel('Count')
plt.legend(title='IBD Status')
plt.show()

# Plot 5: Gluten intolerance prevalence by diet type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='diet_type_sample', hue='gluten_sample')
plt.xticks(rotation=45, ha='right')
plt.title('Gluten Intolerance Prevalence by Diet Type')
plt.xlabel('Diet Type')
plt.ylabel('Count')
plt.legend(title='Gluten Intolerance')
plt.show()
```

```python
# Plot 1: Age distribution by sex without cumulative KDE line
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age_years_sample', hue='sex_sample', multiple="stack", kde=True, kde_kws={'cumulative': False})
plt.title('Age Distribution by Sex')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 1: Age distribution by sex without stacking or cumulative KDE line
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age_years_sample', hue='sex_sample', multiple="layer", kde=True, kde_kws={'cumulative': False})
plt.title('Age Distribution by Sex')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 1: Age distribution by sex with side-by-side histograms and non-cumulative KDE lines
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age_years_sample', hue='sex_sample', multiple="dodge", kde=True, kde_kws={'cumulative': False})
plt.title('Age Distribution by Sex')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 1: Age distribution by sex with side-by-side histograms and non-cumulative KDE lines, saved as SVG
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age_years_sample', hue='sex_sample', multiple="dodge", kde=True, kde_kws={'cumulative': False})
plt.title('Age Distribution by Sex')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')

# Save the plot as an SVG file
plt.savefig("age_distribution_by_sex.svg", format="svg")

# Display the plot
plt.show()
```
